# Basic Definitions

### Business Requirements

What a boss (or customer) really wants. Example "We want more people to sign up"

### Hypothesis

A smart guess you can test. Example: "If we make the form shorter, more people will finish it."

### Design Thinking

Pretend you are the customer, not the boss. Feel their problems first, then think of teh creative ways to fix them.

### Empathy

Seeing the world through their eyes. "Why does this frustrate them?"

###Â Ideation

The "Ideas party" - lots of wild ideas, brainstorming.

### Prototype

A rough sketch or design or model to test your idea

### Test

Let someone try your idea and watch what happens

### AI-Aided Ideation

Asking tools like ChatGPT to join your idea party - to suggest ideas or help write tests faster.

### EDA (Exploratory Data Analysis)

Looking under the hood of your data - drawing pictures and summaries so you understand what is happening
and could share with a non technical person

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
tips.head()

# Compare tip % by day
tips['tip_pct'] = tips['tip'] / tips['total_bill'] * 100
sns.boxplot(x='day', y='tip_pct', data=tips)
plt.title('Tipping Percentage by Day of Week')
plt.show()
```
