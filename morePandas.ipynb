```python
import pandas as pd

# check version
print(pd.__version__)

# Create a dataframe
data = {
    "name": ["Alice", "Bob", "Charlie", "Diana", "Eve"],
    "age": [25, 30, 35, 40, 25],
    "city": ["London", "Paris", "Berlin", "London", "Rome"],
    "salary": [50000, 60000, 75000, 80000, 72000]
}
df = pd.DataFrame(data)
df

```

```python
# Easy CSV handling
df2 = pd.read_csv("penguins.csv")
df2
```

```python
### Explore the data

# first 5 rows
df2.head()

# last 2 rows
df2.tail(2)

# column types + non-null counts
df2.info()

# summary stats (numerical)
df2.describe()

# find out the shape (rows, columns)
df2.shape

# list of the column names
df2.columns
```

```python
# Select a column
df["name"]   # series
df[["name", "city"]]  # multiple cols -> DataFrame

# select rows by index
df.iloc[0]  # first row
df.iloc[:3]  # first 3 rows
df.loc[0, "name"] # value by label

# .iloc = index position
# .loc = label ("name")

# Filter rows
df[df["age"] > 30]  # anyone over 30
df[(df["age"] > 30) & (df["city"] == "Berlin")]
```

```python
# Sorting and Aggregation
# sort
df.sort_values("age")
df.sort_values("salary", ascending=False)

# Aggregations
df["salary"].mean()
df["salary"].max()
df["salary"].min()

# Grouping
df.groupby("city")["salary"].mean()
```
